```python
# Dropping Null Values
```

```python
# Drop rows with null values in the 'URL' column
content = content.dropna(subset=['URL'])
```

```python
# Drop rows with null values in the 'URL' column
reactions = reactions.dropna(subset=['User ID'])
reactions = reactions.dropna(subset=['Type'])
```

```python
# Dropping unnecessary columns
```

```python
content.drop(['Unnamed: 0'], axis=1, inplace=True)
content.drop(['User ID'], axis=1, inplace=True)
content.drop(['URL'], axis=1, inplace=True)
reactions.drop(['Unnamed: 0'], axis=1, inplace=True)
reactions.drop(['User ID'], axis=1, inplace=True)
reaction_type.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# 1. Top 5 Categories by Average Score
top_5_categories_by_score = df.groupby('Category')['Score'].sum().sort_values(ascending=False).head(5).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Category', data=top_5_categories_by_score, palette='viridis')
plt.title('Top 5 Categories by Average Score', fontsize=16)
plt.xlabel('Average Score', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 2. Average Score by Content Type
content_type_by_score = df.groupby('Content_Type')['Score'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Content_Type', data=content_type_by_score, palette='magma')
plt.title('Average Score by Content Type', fontsize=16)
plt.xlabel('Average Score', fontsize=14)
plt.ylabel('Content Type', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 3. Number of Posts Per Month
# Convert 'Datetime' to 'Month' period
df['Month'] = df['Datetime'].dt.to_period('M')

# Group by 'Month' and count 'Content ID'
posts_per_month = df.groupby('Month')['Content ID'].count().reset_index()

# Convert 'Month' to string for plotting
posts_per_month['Month'] = posts_per_month['Month'].astype(str)

# Plotting the data
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Content ID', data=posts_per_month, marker='o', color='blue')
plt.title('Number of Posts Per Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Top 5 Categories by Sentiment and Content Type
top_5_categories_by_sentiment_content = df.groupby(['Category', 'Sentiment', 'Content_Type']).size().reset_index(name='Count')
top_5_categories_by_sentiment_content = top_5_categories_by_sentiment_content.sort_values(by='Count', ascending=False).head(5)

plt.figure(figsize=(12, 8))
sns.catplot(x='Count', y='Category', hue='Sentiment', col='Content_Type', data=top_5_categories_by_sentiment_content, kind='bar', height=5, aspect=1)
plt.suptitle('Top 5 Categories by Sentiment and Content Type', fontsize=16)
plt.tight_layout()
plt.show()
```
